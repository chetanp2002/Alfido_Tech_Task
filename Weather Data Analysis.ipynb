```python
# Extract month name from the date and store in new column named as  month
df['Month']=df.index.month
df['Month']=df['Month'].apply(lambda x: calendar.month_name[x])
df.head()
```

```python
# Get the list of months in which it rained with thunderstom
def checkList(x):
    weatherList=['rain', 'thunderstorm']
    count=0
    for substring in weatherList:
        if substring in x:
            count+=1;
    if(count==2):
        return True
    else:
        return False
print("2) Get the list of months in which it rained with thunderstom\n")
rained = df['Weather'].str.lower().apply(checkList)

month_list = df[rained]['Month'].unique()

print(month_list)
```
